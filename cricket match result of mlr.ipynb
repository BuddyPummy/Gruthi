```python
data_df = data[['Batsman', 'Runs', 'Mins', 'BF', '4s', '6s', 'SR', 'Pos', 'Dismissal','Inns']]
data_df
```

```python
data.fillna(data.mean())
data['Mins'].fillna(data['Mins'].mean(),inplace=True)
data
```

```python
cat_features = ['Dismissal']
data_df = pd.get_dummies(data_df,columns=cat_features)
print(data_df.columns)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().fit(data_df[['4s','6s','SR']]) 

data_normalized = scaler.transform(data_df[['4s','6s','SR']])
data_normalized
```

```python
data_df_normalized=pd.DataFrame(data_normalized,columns=['4s','6s','SR'])
data_df_normalized=data_df_normalized.join(data_df[data_df.columns.drop(['4s','6s','SR'])])
data_df_normalized
```

```python
Y = data_df_normalized["Dismissal_bowled"]
X = data_df_normalized[data_df_normalized.columns.drop("Dismissal_bowled")]
print(X.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2, random_state=100)
X_train.shape , X_test.shape
```

```python

from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train,Y_train)

train_score=model.score(X_train,Y_train)
test_score=model.score(X_test,Y_test)
print('Train Score (R-Squared): ',train_score)
print('Test Score (R-Squared)',test_score)
```

```python
from sklearn.metrics import mean_squared_error

train_predictions = model.predict(X_train)
train_RMSE=mean_squared_error(Y_train,train_predictions)**0.5

test_predictions = model.predict(X_test)
test_RMSE=mean_squared_error(Y_test,test_predictions)**0.5
print('Train RMSE : ',train_RMSE)
print('Test RMSE : ',test_RMSE)
```
