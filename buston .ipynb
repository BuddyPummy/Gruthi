```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('D:/aiml/boston_housing.csv')
df
```

```python
MEDV=df.iloc[:,0].values
AGE=df.iloc[:,1].values
MEDV,AGE
```

```python
plt.scatter(MEDV,AGE)
```

```python
regressor=LinearRegression()
regressor.fit(MEDV.reshape(-1,1),AGE)
```

```python
print(regressor.coef_)
print(regressor.intercept_)
```

```python
AGE_pred=regressor.predict(AGE.reshape(-1,1))
AGE_pred
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(AGE,AGE_pred)
```
